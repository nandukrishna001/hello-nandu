```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/NCI60_X (1).csv",index_col=0)
```

```python
data=data.drop_duplicates()
data.shape
```

```python
data.columns
```

```python
x=data
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters = 5, n_init = 20)
```

```python
km.fit(x)
```

```python
cluster_labels = km.predict(x)
```

```python
cluster_labels
```

```python
pd.Series(cluster_labels).value_counts()
```

```python
x
```

```python
km.inertia_
```

```python
inertia_list = []


for i in range(1,11):
  km = KMeans(n_clusters=i, n_init = 20)
  km.fit(x)
  inertia = km.inertia_
  inertia_list.append(inertia)
```

```python
inertia_list
```

```python
## elbow method
```

```python
plt.plot(range(1,11), inertia_list)
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.show();
```

```python
from sklearn.preprocessing import scale
```

```python
x = pd.DataFrame(scale(data), index = data.index, columns = data.columns)
```

```python
from sklearn.decomposition import PCA
```

```python
pc=PCA(n_components=4)
```

```python
pc_array=pc.fit_transform(x)
```

```python
pc_array.shape
```

```python
pc_data=pd.DataFrame(pc_array,columns=["pc1","pc2","pc3","pc4"])
```

```python
pc_data.shape
```

```python
pc_data
```

```python
pc_data.corr().round(1)
```

```python
pc.explained_variance_
```

```python
#### 2.5 is the higest variance, 788.13 second highest variance,468.77  is the third 399.08 and fourth 294.70
```

```python
pc.explained_variance_ratio_  * 100
```

```python
### 1 PC is containing 788 percent of the data , 2 PC is containing 468.77 percent and 3pc is containig 399,08 of the data and so on.
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4], pc.explained_variance_ratio_  * 100, "-o")
plt.show()
```
